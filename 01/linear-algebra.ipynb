```python
def vect_vect_multiplication(u, v):
    """
    A function that does vector x vector multiplication in the following way:
    1. We verify if both vectors have the same shape. u.shape = (4,) v.shape = (4,)
    2. The for loop sums the element-wise products
    """
    assert u.shape[0] == v.shape[0]

    n = u.shape[0]

    result = 0.0

    for i in range(n):
        result = result + u[i] * v[i]

    return result
    
```

```python
vect_vect_multiplication(u, v)
```

```python
u.dot(v)
```

## Matrix-vector multipication

```python
U = np.array([
    [1, 4, 7, 9],
    [2, 6, 3, 8],
    [3, 4, 0, 4]
])
```

```python
U
```

```python
U.shape
```

```python
def matrix_vector_multiplication(U, v):
    """ 
    1. Check the dimensionality match; the number of columns in the matrix should match the number of rows in the vector.
    2. Initialise the resulting matrix with zeros with the same number of rows as the matrix U.
    3. For each element i in the result vector, we calculate the dot product of element i of matrix U and the vector v.
    """
    assert U.shape[1] == v.shape[0]

    num_rows= U.shape[0]

    result = np.zeros(num_rows)

    for i in range(num_rows):
        result[i] = U[i].dot(v)

    return result
```

```python
matrix_vector_multiplication(U, v)
```

```python
U.dot(v)
```

```python
V = np.array([
    [1, 1, 2],
    [0, 0.5, 1],
    [0, 2, 1],
    [2, 1, 0],
])
```

```python
def matrix_matrix_multi(U,V):
    
    assert U.shape[1]==V.shape[0]

    num_rows = U.shape[0]

    num_cols = V.shape[1]

    result = np.zeros((num_rows, num_cols))

    for i in range(num_cols):
        vi = V[:,i]
        Uvi= matrix_vector_multiplication(U, vi)
        result[:,i]= Uvi
    return result
```

```python
matrix_matrix_multi(U,V)
```

```python
U.dot(V)
```
